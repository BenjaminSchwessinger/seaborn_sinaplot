## Sinaplots:

```python
from matplotlib import pyplot as plt
import seaborn as sns
from sinaplot import sinaplot


%matplotlib inline
sns.set(font='Arial')
plt.rcParams['svg.fonttype'] = 'none'
style = sns.axes_style('white')
style.update(sns.axes_style('ticks'))
style['xtick.major.size'] = 2
style['ytick.major.size'] = 2
sns.set(font_scale=2, style=style)
pal = sns.color_palette(['#0072b2', '#d55e00', '#009e73', '#f0e442', '#cc79a7'])
sns.set_palette(pal)
sns.palplot(pal)
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

Sinaplots are simply stripplots where the jitter applied is sampled randomly from the uniform distribution, but then scaled to the width of the kernel density estimate for the data at the position of that point.

`matplotlib_sinaplot` contains one function that produces sinaplots using the `seaborn` and `matplotlib` libraries. Using it should be almost identical to using `seaborn.violinplot`: 

```python
fig, ax = plt.subplots(figsize=(8, 5))
sinaplot(
    x='sex', y='age',
    hue='survived', data=titanic,
    violin=False
)
plt.show()
```

With sinaplots, by default the width of the kde jitter is scaled to the number of examples, but this can be set back to the violinplot default of same area for all kdes:

```python
fig, ax = plt.subplots(figsize=(8, 5))
sinaplot(
    x='sex', y='age',
    hue='survived', data=titanic,
    violin=False, scale='area',
)
plt.show()
```

As with violinplots we can produce split sinaplots when there are only two levels

```python
fig, ax = plt.subplots(figsize=(8, 5))
sinaplot(
    x='sex', y='age',
    hue='survived', data=titanic,
    split=True, point_facealpha=1,
    violin=False
)
plt.show()
```

We can also show the underlying violin or add summary stats

```python
fig, ax = plt.subplots(figsize=(8, 5))
sinaplot(
    x='sex', y='age',
    hue='survived', data=titanic,
    split=True,
    point_facealpha=1, violin_facealpha=0.2,
    violin=True, inner='quart',
)
plt.show()
```
